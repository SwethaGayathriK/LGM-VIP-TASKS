```python
song_data = train[['name','target']]
song_data
```

```python
song_data = song_data.groupby('name').sum()
song_data.sort_values(by=['target'], inplace=True, ascending=False)
song_data = song_data.reset_index() 
song_data.head(10)
```

```python
# function to recommend new songs
def Music_Recommendation(song_name, k):
    # to convert dataframe into array
    song_list = (song_data['name']).to_list() 
    # to check if song selected by user is available?
    if song_name in song_list :
        print("name"+"'"+song_name+"'"+" is available")
        song_index = (song_list.index(song_name))
        # total listen count of the song which is selected by user
        song_listen_count = (song_data['target'].iloc[song_index])
        new_distance_list = []
        for i in range(len(song_data)):
            if i != song_index :
                # calculate euclidean distance
                abs_distance = abs(song_listen_count - song_data['target'][i])
                new_distance_list.append([abs_distance,song_data['name'][i]])
        # to sort distance list by ascending order
        new_distance_list = sorted(new_distance_list)
        print("Top" + str(k) + "Recommendation for you.......")
        for i in range(k):
            print(i+1,")",new_distance_list[i][1])
    else:
        print("Song"+"'"+song_name+"'"+"is not available")
        print("Top" + str(k) + "Recommendation for you.......")
        recommend_song = (song_data['name']).head(k)
        for i in range(k):
            print(i+1,")",recommend_song['name'][i])
k = 10
Music_Recommendation('FLY OUT', k)
```
